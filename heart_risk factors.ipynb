CHI- SQUARE TEST FOR ASSOCIATION
Steps for hypothesis testing:
>formulate the statement of hypothesis
>state the significance level
>test criteria
>decision-making
>conclusion

```python
# contingency table
cp_targ=pd.crosstab(df['target'],df['cp'])
display(cp_targ)
sns.heatmap(data=cp_targ,annot=True,cmap='magma',fmt='d')
plt.title('Heatmap for the distribution')
plt.show()
```

```python
# Hypothesis statement
print("Statement of Hypothesis")
print('H0: No association between the variable under test')
print('Ha: There is an association between the variable under test \n')
alpha=float(input("Enter alpha value as a decimal value:"))
print(f'tabulated pvalue: {alpha} \n')
# Test statistic
chi_val,pval,dof,e_freq=chi2_contingency(cp_targ)
print(f'Test statistic: {chi_val}') 
print(f"pvalue: {pval}")
print(f'degree of freedom: {dof}')
print(f"Expected frequency array:\n {e_freq} \n")
#decision and conclusion
if (pval > alpha):
    print('Decision')
    print('fail to reject null hypothesis \n')
    print('Conclusion')
    print('No association between cp and heart disease occurence')
elif (pval == alpha):
    print('Decision')
    print('fail to reject null hypothesis \n')
    print('Conclusion')
    print('No association between cp and heart disease occurence') 
else:
    print('Decision')
    print('reject null hypothesis \n')
    print('Conclusion')
    print('there is an association between cp and heart disease occurence')
```

```python
exang_targ=pd.crosstab(df['target'],df['exang'])

# Test statistic
chi_val,pval,dof,e_freq=chi2_contingency(exang_targ)
print(f'Test statistic: {chi_val}') 
print(f"pvalue: {pval}")
print(f'degree of freedom: {dof}')
print(f"Expected frequency array:\n {e_freq} \n")
alpha=int(input("Enter percentage value for alpha: "))
alpha=alpha*1/100
print(f'tabulated pvalue: {alpha} \n')
#decision and conclusion
if (pval > alpha):
    print('Decision')
    print('fail to reject null hypothesis \n')
    print('Conclusion')
    print('No association between exercise-induced angina and heart disease occurence')
elif (pval == alpha):
    print('Decision')
    print('fail to reject null hypothesis \n')
    print('Conclusion')
    print('No association between exercise-induced angina and heart disease occurence') 
else:
    print('Decision')
    print('reject null hypothesis \n')
    print('Conclusion')
    print('there is an association between exercise-induced angina and heart disease occurence')
```

TEST OF SIGNIFICANCE FOR CHOLESTROL VARIABLE 
 Hypothesis statement:
    H0: there's no significance difference in cholestrol levels for occurence of Heart disease 
    Ha: there's a significant difference in occurence of heart disease in a person
 Level of significance:
    test at 1% level of significance

```python
s_data=df[['chol','target']]
print('level target:',s_data['target'].unique())
first_group=s_data[s_data['target']==0]['chol']
second_group=s_data[s_data['target']==1]['chol']
statistic_val,p_val=ztest(first_group,second_group,alternative='two-sided')
print(f'Test statistic value: {statistic_val}')
print(f'P-Value: {p_val}')

def wald_test():
    alpha=int(input('PLEASE ENTER YOUR PERCENTAGE LEVEL OF SIGNIFICANCE: '))
    alpha=alpha/100
    print('alpha:',alpha)
    if p_val == alpha:
        print('DECISION')
        print('there is no enough evidence that suffices to reject the null hypothesis \n')
        print('CONCLUSION')
        Print('cholestrol levels have a less significant effect in determining the occurence of heart disease in a person')
    elif p_val < alpha:
        print('DECISION')
        print('reject the Null hypothesis \n')
        print('CONCLUSION')
        print('cholestrol variable has a significance effect in occurence of heart disease')
    else:
        print('DECISION')
        print('there is no enough evidence that suffices to reject the null hypothesis \n')
        print('CONCLUSION')
        print('cholestrol levels have a less significant effect in determining the occurence of heart disease in a person')
wald_test()
```

```python
f_data=df[['ca','age']]
print(f_data.shape)
f_data['ca'].unique()
#hypothesis
H0="All group means are equal"
Ha="Atleast one group differ significantly"
print(f">>HYPOTHESIS STATEMENT \n H0:{H0}\n Ha:{Ha} \n")
#level of significance
lof=int(input('enter the level of significance as a numeric: '))
lof=lof/100
print(f'>>SIGNIFICANCE LEVEL \n test of significance ={lof} \n')
#test statistic
print('>>test statistic'.upper())
groups=[f_data[f_data['ca']==i]['age']for i in f_data['ca'].unique()]
fval,pval=f_oneway(*groups)
print(f'test statistic value: {fval}')
print('P-value:', pval,'\n')
#function defition for decision and conclusion
def f_fn():
    if (pval > lof):
        print('>>decision'.upper())
        print('Do not reject the null hypothesis \n')
        print('>>conclusion'.upper())
        print(H0)
    elif (pval == lof):
        print('>>decision'.upper())
        print('Do not reject the null hypothesis \n')
        print('>>conclusion'.upper())
        print(H0)
    else:
        print('>>decision'.upper())
        print('Reject the null hypothesis \n')
        print('>>conclusion'.upper())
        print(Ha)
f_fn()
```

ML MODEL- SUPERVISED ALGORITHM (KNN)
> FEATURES SELECTION
> SCALING
> TRAINING MODEL
> PREDICTING
> METRICS

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,roc_auc_score
```

```python
#feature and target column
y=df['target']
x=df.drop('target',axis=1)
#training model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#scaling feature
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
#model training 
knn=KNeighborsClassifier(n_neighbors=3)
#model fit
knn.fit(x_train,y_train)
#model prediction
y_classify=knn.predict(x_test)
#metrics
accuracy=accuracy_score(y_test,y_classify)
print(f'accuracy score; {accuracy:.4f}')
conf_matrix=confusion_matrix(y_test,y_classify)
print(f'\n confusion_matrix: \n {conf_matrix} \n')
#plot of confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
class_report=classification_report(y_test,y_classify)
print(f'\n classification report: \n {class_report}')
prec_score=precision_score(y_test,y_classify)
print(f'precision score: {prec_score} \n')
ras = roc_auc_score(y_test, y_classify)
print('ROC Score:', ras)
```
