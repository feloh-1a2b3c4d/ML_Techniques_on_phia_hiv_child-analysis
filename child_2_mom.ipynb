```python
from scipy.stats import (ttest_ind, f_oneway, chi2_contingency)
from statsmodels.stats.weightstats import ztest
print('TEST 1. CHI-SQUARE TEST FOR ASSOCIATION BETWEEN MOM SICK AND CHILD STATUS')
#contingency table creation
cont_data=d_frame.dropna(subset=['hivstatusfinal', 'momsick'])
cont_table=pd.crosstab(cont_data['hivstatusfinal'],cont_data['momsick'])
plt.figure(figsize=(7,7))
plt.title("contingency heatmap",color='green')
plt.xlabel("mom sick",color='blue')
plt.ylabel('hiv status of child',color='brown')
sns.heatmap(cont_table,annot=True,cmap='viridis',fmt='.2f')
plt.show()

print('1. CHI-SQUARE TEST STATISTIC \n')
chi_val,p_value,dof,expected_freq=chi2_contingency(cont_table)
print(f'Chi-square statistic: {chi_val}')
print(f'p-value: {p_value}')
print(f'Degrees of freedom: {dof}')
print('Expected frequencies:')
print(expected_freq)

print('\n2. HYPOTHESIS TESTING \n')
alpha=0.05
def chi_square():
    if p_value > alpha:
        print('DECISION; ')
        print("Do not reject Null hypothesis \n")
        print('CONCLUSION ; ')
        print(' there is no association between mom being sick and the child hiv status')
    elif p_value == alpha:
        print('DECISION; ')
        print("Do not reject Null hypothesis \n")
        print('CONCLUSION ; ')
        print(' there is no association between mom being sick and the child hiv status')
    else :
        print('DECISION; ')
        print("Reject Null hypothesis \n")
        print('CONCLUSION ;')
        print(' there is an association between mom being sick and the child hiv status')
chi_square()
```

```python
print(f'TEST 2. Z_TEST FOR SIGNIFICANCE OF PEDTRI90')# we use ztest to ttest since our sample is large
print(f'\n 1. Z_TEST STATISTIC \n')
z_data=d_frame.dropna(subset=[ 'pedtri90', 'hivstatusfinal'])
levels_hiv=z_data['hivstatusfinal'].unique()
print(f'levels for hiv; \n {levels_hiv}')
group1=z_data[z_data['hivstatusfinal']==1.]['pedtri90']
group2=z_data[z_data['hivstatusfinal']==2.]['pedtri90']
z_value,p_zvalue=ztest(group1,group2,alternative="two-sided")
print(f'z_statistic; {z_value}\n p_value; {p_zvalue}')
print(f'\n 2. HYPOTHESIS TESTING')
alpha=0.05
def z_statistic():
    print('\n STATEMENT OF HYPOTHESIS')
    print("H0 ; hypothesis of no difference")
    print('Ha ; Hypothesis of difference \n')
    if p_zvalue > alpha:
        print('DECISION')
        print("Do not reject the Null hypothesis \n")
        print("CONCLUSION")
        print('there is no significant difference between the child status hiv ')
    elif p_zvalue==alpha:
        print('DECISION')
        print("Do not reject the Null hypothesis \n")
        print("CONCLUSION")
        print('there is no significant difference between the child status hiv ')
    else :
        print('DECISION')
        print("Reject the Null hypothesis \n")
        print("CONCLUSION")
        print('there is a significant difference between the child status hiv ')
z_statistic()
```

```python
print('TEST 3. ONE-WAY ANOVA FOR WEALTHQUANTILE AND AGE OF CHILD  \n')
print('1. F_ONEWAY TEST CRITERIA \n')
f_data=d_frame.dropna(subset=['wealthquintile', 'age_x'])
levels_wealth=f_data['wealthquintile'].unique()
print(f"levels; {levels_wealth}")
groups=[f_data[f_data['wealthquintile']==lev]['age_x'] for lev in f_data['wealthquintile'].unique()]
f_val,p_fval=f_oneway(*groups)
print(f'f_statistic; {f_val}\n p_value; {p_fval}')
print(f'\n 2. HYPOTHESIS TESTING')
alpha=0.05
def f_statistic():
    print('\n STATEMENT OF HYPOTHESIS')
    print("H0 ; hypothesis of no difference")
    print('Ha ; Hypothesis of difference \n')
    if p_zvalue > alpha:
        print('DECISION')
        print("Do not reject the Null hypothesis \n")
        print("CONCLUSION")
        print('there is no significant difference between group means ')
    elif p_zvalue==alpha:
        print('DECISION')
        print("Do not reject the Null hypothesis \n")
        print("CONCLUSION")
        print('there is no significant difference between group means ')
    else :
        print('DECISION')
        print("Reject the Null hypothesis \n")
        print("CONCLUSION")
        print('Atleast one group mean differ significantly ')
f_statistic()
```

```python
print(f"TEST 4. T_TEST FOR SIGNIFICANCE OF PARENTGUARDQX")
print(d_frame['parentguardqx'].unique())
print('\n 1. HYPOTHESIS TESTING\n')
print("H0: There is no significant difference in child HIV status")
print('H1: There is a significant difference in the child HIV status \n')
print('2. TEST STATISTIC \n')
t_data=d_frame.dropna(subset=['parentguardqx', 'hivstatusfinal'])
g1=t_data[t_data['hivstatusfinal']==1.]['parentguardqx']
g2=t_data[t_data['hivstatusfinal']==2.]['parentguardqx']
test_statistic,p_tval=ttest_ind(g1,g2)
print(f'T statistic: {test_statistic}')
print(f'P-Value: {p_tval} \n')
def t_test():
    alpha = 0.05
    if p_tval == alpha:
        print('DECISION')
        print('We fail to reject the Nul hypothesis \n')
        print('CONCLUSION')
        print('There is no significant difference in child HIV status')
    elif p_tval > alpha:
        print('DECISION')
        print('We fail to reject the Null Hypothesis \n')
        print('CONCLUSION')
        print('there is no significant difference in child HIV status')
    else:
        print('DECISION')
        print('reject the Null hypothesis \n')
        print('CONCLUSION')
        print('There is a significant difference in the child HIV status')
t_test()
```

```python
print('MACHINE LEARNING')
print('\n1. DATA SELECTION FEATURE AND TARGET COLUMN \n')
target_col=d_frame["hivstatusfinal"]
feature_col=d_frame.drop('hivstatusfinal',axis=1)
display(target_col.head())
print('target column; ',target_col.shape)
display(feature_col.head())
print('features column; ',feature_col.shape)

print('\n2. DATA SPITTING INTO TRAINNING AND TEST \n')
X_train,X_test,y_train,y_test=train_test_split(feature_col,target_col,test_size=0.3,random_state=42)
print("x_train shape:", X_train.shape)
print("x_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape", y_test.shape)

print('\n3. DATA NORMALIZATION \n')
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
print('scaling on data to reduce cases of overfitting')

print('\n MODEL TRAINING AND FITTING')
lr_model=LogisticRegression(max_iter=1000)
rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
svm_model=SVC(kernel='rbf',C=1,gamma='scale')
knn_model=KNeighborsClassifier(n_neighbors=3)
#training models on training sets
lr_model.fit(X_train,y_train)
rf_model.fit(X_train,y_train)
svm_model.fit(X_train,y_train)
knn_model.fit(X_train,y_train)
# models prediction basing on features test set
ypred_lr=lr_model.predict(X_test)
ypred_rf=rf_model.predict(X_test)
ypred_svm=svm_model.predict(X_test)
ypred_knn=knn_model.predict(X_test)
```
